```python
import pandas as pd
import numpy as np

#import data#
df = pd.read_excel (r'D:\Data Science\Python\Github\Supervised Machine learning\Decision Tree_Data.xlsx')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
df['company_n'] = le_company.fit_transform(df['company'])
df['job_n'] = le_job.fit_transform(df['job'])
df['degree_n'] = le_degree.fit_transform(df['degree'])
```

```python
df
```

```python
## Define Independent Variables 
Factors=df.loc[:,'company_n':'degree_n']
Factors
```

```python
Target=df['salary_more_then_100k']
from sklearn import tree
model = tree.DecisionTreeClassifier()
Target=df['salary_more_then_100k']
model.fit(Factors,Target)
```
